### Plot and Truncate
- Truncate based on a simple pressure threshold (pressure > 0.3 dbar appears to be in water)
- Plot conductivity profiles

```python
fig, axs = plt.subplots(figsize = (10, 10))

for filename in filelist_sorted:
    print(filename)

    df_full = pd.read_csv(filename, header = 28)
    
    # Truncate based on a simple pressure threshold (pressure > 0.3 dbar appears to be in water)
    df_inwater = df_full[df_full['Pressure (Decibar)'] > 0.3]
    
    axs.plot(df_inwater['Conductivity (MicroSiemens per Centimeter)'], df_inwater['Pressure (Decibar)'])

axs.set_ylabel('Pressure (Decibar)')
axs.set_xlabel('Conductivity (S/cm)')
plt.gca().invert_yaxis() # Flip the y-axis

plt.savefig('TownCreek_Profiles.png', dpi = 300)
```

### Create large df of all dfs
- NB: location is arbitrary currently
- Truncate based on a simple pressure threshold (pressure > 0.3 dbar appears to be in water)
- Use gsw/TEOS-10 to calculate salinity from cond, temp, press

```python
dist = 0

# Create empty DF to fill iteratively
df_large = pd.DataFrame([])

for filename in filelist_sorted:
    print(filename)
    df_individ = pd.read_csv(filename, header = 28)
    
    p = df_individ['Pressure (Decibar)']
    print(np.nanmax(p))
    
    # Only for full profiles
    if np.nanmax(p) > 2:
        # Add a contrived (lazy) distance for now; use file's lat/lon later
        df_individ.loc[:, 'Distance (contrived)'] = np.ones(len(df_individ))*dist

        # Truncate based on a simple pressure threshold (pressure > 0.3 dbar appears to be in water)
        df_inwater = df_individ[df_individ['Pressure (Decibar)'] > 0.3]
        
        # Extract key parameters
        C_mScm = df_inwater['Conductivity (MicroSiemens per Centimeter)']/1000
        t = df_inwater['Temperature (Celsius)']
        p = df_inwater['Pressure (Decibar)']

        # Calculate salinity with TEOS-10 module (PSS-78 algorithm)
        df_inwater['Salinity'] = gsw.conversions.SP_from_C(C_mScm, t, p)

        df_large = pd.concat([df_large, df_inwater])

        dist += 1 # increment contrived distance
     
    # Reset vars for next loop
    C_mScm = []
    t = []
    p = []
    
df_large
```

```python
n_grid = 100

salinity = df_large['Salinity']
distance = df_large['Distance (contrived)']
depth = df_large['Pressure (Decibar)']

plotVar = salinity

dist_vec_unique = np.linspace(np.min(distance), np.max(distance), n_grid) # just for gridding
depth_vec_unique = np.linspace(np.min(depth), np.max(depth), n_grid) # just for gridding

dist_grid, depth_grid = np.meshgrid(dist_vec_unique, depth_vec_unique)
plotVar_grid = griddata((distance, depth), plotVar, (dist_grid, depth_grid), method = 'linear')
```

```python
fig, axs = plt.subplots(figsize = (10, 10))
im = plt.contourf(dist_grid, depth_grid, plotVar_grid)#, levels = 40)#, levels = cvec, cmap=myCM, extend="both"))

sc = plt.scatter(distance, depth, c = 'black', s = 5, alpha = 1)
#                cmap=myCM, alpha = 1, vmin=vmn, vmax=vmx)

axs.set_ylabel('Pressure (Decibar)')
axs.set_xlabel('Distance (contrived)')
plt.gca().invert_yaxis() # Flip the y-axis
plt.gca().invert_xaxis() # Flip the x-axis for now, since Cape Fear River is W edge

cbar = fig.colorbar(im)
cbar.ax.set_ylabel('Salinity (PSU)')

# plt.savefig('TownCreek_SalinityProfiles.png', dpi = 300)
```
